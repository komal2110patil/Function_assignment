```python
 #Q4]write a python that checks if a given number is prime or not from 1 to 200?
 def is_prime(number):
    if number <= 1:
        return False
    if number <= 3:
        return True
    if number % 2 == 0 or number % 3 == 0:
        return False
    i = 5
    while i * i <= number:
        if number % i == 0 or number % (i + 2) == 0:
            return False
        i += 6
    return True
```

```python
#5]create an iterator class that generate fibonacci sequence upto specified number of terms?
class Fibonacciiterator:
  def __init__(self, limit):
    self.limit = limit
    self.a = 0
    self.b = 1
    self.count = 0

    def __iter__(self):
       return self

  def __next__(self):
    if self.count >= self.limit:
      raise StopIteration
```

```python
#6]write a generator function in python that yields the power of 2 upto a given exponent?
def power_of_two_generator(exponent):
    for i in range(exponent + 1):
        yield 2 ** i
```

```python
#7]implement a generator function that reads a file line by line and yield each line as a string?
def read_file_lines(file_path):

  try:
    with open(file_path, 'r') as file:
        for line in file:
            yield line
  except FileNotFoundError:
    print(f"File '{file_path}' not found.")
```

```python
#9]write a python programme that uses 'map()' to convert list of tempratures from celcius to Fahrenheit?
def celsius_to_Fahrenheit(celsius):
  return (celsius * 9/5) + 32

celsius_temperatures = [0, 10, 20, 30, 40]

fahrenheit_temperatures = list(map(celsius_to_Fahrenheit, celsius_temperatures))
print(fahrenheit_temperatures)
```

```python
#10]create a python programme that uses filer to remove all the vowels from a given string?
def remove_vowels(string):
  vowels = 'aeiouAEIOU'
  return ''.join(filter(lambda x: x not in vowels, string))

```
